```python
# Make HTTP requests
import requests
# Scrape data from an HTML document
from bs4 import BeautifulSoup
# I/O
import os

import re
```

```python
#this is a wikipedia page with links to lists of artists by genre
linksURL = "https://en.wikipedia.org/wiki/Lists_of_musicians"


page = requests.get(linksURL)
html = BeautifulSoup(page.text, 'html.parser')
div = html.find_all('div', class_='div-col')

links = []
genres = []

for d in div:

    a_elements = d.find_all('a')


    hrefs = [a['href'] for a in a_elements]
    titles = [a.get('title') for a in a_elements]

    links.extend(hrefs)
    genres.extend(titles)

genres = [s.replace("List of ", "") for s in genres]


```

```python
#i only want to go to hip hop
i = genres.index("hip hop artists")

genrelinks = dict(zip(genres[0:211],links[0:211]))
genrelinks
```

```python
baseurl = "https://en.m.wikipedia.org"

artist_names = dict()

for genre,link in genrelinks.items():
    newurl = baseurl + link

    page = requests.get(newurl)
    html = BeautifulSoup(page.text, 'html.parser')

    mainbody = html.find("div",class_="mw-parser-output")

    lis = mainbody.select('li:not([id]):not([class])')

    names = []
    for li in lis:
        
        a = li.find_all('a')

        t = [w.get('title') for w in a]

        names.extend(t)
    
    artist_names[genre] = list(filter(lambda x: x is not None, names))



artist_names
    
```

```python
for key, values in artist_names.items():
    file_name = "complete_artists/" + key + '.txt'
    with open(file_name, 'w+') as file:
        for value in values:
            file.write(value + '\n')
    print(f"File '{file_name}' created successfully.")
```
